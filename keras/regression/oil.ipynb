```python
import pandas as pd
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
import numpy as np
from keras.datasets import boston_housing
from keras.models import Sequential
from keras.layers import Dense
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
%matplotlib inline
```

```python
x = pd.read_csv('Inputs.txt', header=None, delimiter=r"\s+").values
y = pd.read_csv('Targets8.txt', header=None, delimiter=r"\s+").values
```

```python
flags = ~np.isnan(y[:, 0])
y = y[flags]
x = x[flags]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)
```

```python
x_valid, x_test, y_valid, y_test = train_test_split(x_test, y_test, test_size=0.5, random_state=42)
```

```python
scaler = StandardScaler().fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
x_valid = scaler.transform(x_valid)
```

```python
model = Sequential()
model.add(Dense(1024, input_dim=24, kernel_initializer='normal', activation='relu'))
model.add(Dense(512, kernel_initializer='normal', activation='relu'))
model.add(Dense(256, kernel_initializer='normal', activation='relu'))
model.add(Dense(128, kernel_initializer='normal', activation='relu'))
model.add(Dense(1, kernel_initializer='normal'))
# For a mean squared error regression problem
model.compile(optimizer='adam', loss='mse')
model.fit(x_train, y_train, epochs=10000, batch_size=20000, validation_data=(x_valid, y_valid))
```

```python
score = model.evaluate(x_test, y_test, batch_size=128)
print(score)
```

```python
y_test_predict = model.predict(x_test)
#y_train_predict = model.predict(x_train)
```

```python
fig = plt.gcf()
fig.set_size_inches(5, 5)
plt.scatter(y_test, y_test_predict)
plt.xlim(0, 100)
plt.ylim(0, 100)
```
